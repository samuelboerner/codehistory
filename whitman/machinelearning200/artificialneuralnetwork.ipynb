**Step 1: Importing Libraries and Dataset**

```python
# Libraries
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout

from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/My Drive/2022-23/S23/CS-200/Week 10/artificialneuralnetwork/

data = pd.read_csv('data.csv')
```

**Step 2: Extracting Independent and Dependent Variables**

```python
# Splitting
X = data.iloc[:, 2:32].values
y = data.iloc[:, 1].values
```

**Step 3: Feature Engineering: Encoding**

```python
# Label Encoding Categorical Data
labelencoder_Y = LabelEncoder()
y = labelencoder_Y.fit_transform(y)
```

**Step 4: Split into Train and Test Sets**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

**Step 5: Feature Engineering: Scaling**

```python
scale_X = StandardScaler()
X_train = scale_X.fit_transform(X_train)
X_test = scale_X.transform(X_test)
```

**Step 6: Initialize the ANN**

```python
classifier = Sequential()
```

**Step 7: Adding Input and Hidden Layer**

```python
# First Layer
classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu', input_dim = 30))
```

**Step 8: Adding Second and Third Hidden Layer**

```python
classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu'))
```

**Step 9: Adding Output Layer**

```python
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Dropout to prevent overfitting
classifier.add(Dropout(rate = 0.1))
```

**Step 10: Compiling the Model**

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

**Step 11: Fitting**

```python
classifier.fit(X_train, y_train, batch_size = 100, epochs = 150)
```

**Step 12: Predict using Test Set**

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > .5)
```

**Step 13: Making Confusion Matrix**

```python
cm = confusion_matrix(y_test, y_pred)
```

**Step 14: Calculate Metrics**

```python
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy is {}'.format(accuracy))

# Precision
precision = precision_score(y_test, y_pred)
print('Accuracy is {}'.format(precision))

# Recall
recall = recall_score(y_test, y_pred)
print('Accuracy is {}'.format(recall))

# Recall
f1 = f1_score(y_test, y_pred)
print('F-measure is {}'.format(f1))
```
