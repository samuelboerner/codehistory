**Step 1: Import Libraries and Dataset**

```python
import numpy as np
import pandas as pd

%pip install apriori_python
from apriori_python import apriori

from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/My Drive/2022-23/S23/CS-200/Week 16/associationmining/

data = pd.read_csv('Dataset.csv', header = None)
```

**Step 2: Preparing Transactions**

```python
transactions = []
for i in range(0, 25):
  transactions.append([str(data.values[i, j]) for j in range(0, 5)])
```

**Step 3: Building the Model**

```python
freq_items, rules = apriori(transactions, minSup = 0.249, minConf = 0.699)
```

**Step 4: Interpreting Results**

```python
# First result from ruleset
print(rules[0])
```

But, you can write a function to calculate the support for a specific item in the list of transactions, and then divide a rule's confidence by the support of the consequent. L(A->B) = C(A->B)/S(B)

```python
def support(transactions, item):
  # Sum count
  count = 0
  for transaction in transactions:
      if item in transaction:
          count += 1

  # Calculate support
  support = count / len(transactions)
  return support
```
